# 1.3 A Simple Environmental Model

```python
%%matlab
alpha = 1;
c0 = 1;
t = [0:0.1:1];
f = c0*exp(alpha*t);
plot(t,f)
```

```python
%pylab inline
alpha = 1
c0 = 1
# t = [0:0.1:1]
t = linspace(0,1,10)
f = c0*exp(alpha*t)
plot(t,f)
```

```python
%%matlab
lambda = 1; 
c0 = 10;
t = [0:0.1:1];
f = c0*exp(-lambda*t);
plot(t,f)
grid;
```

```python
%pylab inline
lam = 1; 
c0 = 10;
t = linspace(0,1,10)
f = c0*exp(-lam*t);
plot(t,f)
grid();
```
