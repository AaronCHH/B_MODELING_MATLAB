## 13.2 1D Implementation

```python
%%matlab
h0 = 10;
Qx0 = 5.e-5
K = 0.0001;

x = [1:1:100];

hc = h0 + (Qx0/(K*h0))*x;
hu = sqrt(h0*h0 + (2*Qx0/K)*x);

plot(x, hc, '-b', x, hu, '--g');
legend('confined', 'unconfined', 2)
```

```python
%%matlab
L = 100;
h1 = 12;
hc = h0 + ((h1-h0)/L)*x;
hu = sqrt(h0*h0 + ((h1*h1-h0*h0)/L)*x);
plot(x,hc,'--g',x,hu,'-b');
legend('confined','unconfined',2)
```

## 13.3 2D Implementation

```python
%%matlab
[x,y] = meshgrid([0:10:100],[10:2:20]);

hc = h0 + (Qx0/(K*h0))*x;
hu = sqrt (h0*h0 + (2*Qx0/K)*x);

surf (x,y,hc);
hold on
surf (x,y,hu);
```
